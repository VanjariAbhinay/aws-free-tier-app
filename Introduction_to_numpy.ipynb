```python
import time
```

##**Python List**

###**Calculating the square of each element**

```python
start = time.time()

list1 = list(range(100_000_000))

for i in list1:
  i*=i

end = time.time()
list_time = end - start

print(f"Time taken using List: {list_time:.6f} seconds")
```

##**NumPy Arrays**

###**Calculating the square of each element**

```python
import numpy as np

start = time.time()

numpy_array = np.arange(100_000_000)
numpy_array = numpy_array**2

end = time.time()
numpy_time = end - start

print(f"Time taken using NumPy: {numpy_time:.6f} seconds")
```

##**Speedup Calculation:Lists vs. Numpy Arrays**

```python
speedup = list_time / numpy_time
```

```python
speedup
```

# NumPy
- Short for Numerical Python
- Foundational Package for Numerical computing in python
- Provides an ndarray object which is an efficient multidimensional array providing fast array oriented operations
- Provides a lot of functions

##**1D Array Creation Methods**

###**From List**

You can create a NumPy array directly from a Python list using np.array()

*syntax* : np.arange([list of elements])

```python
np.array([1,2,3,4])
```

###**Using built-in methods**

NumPy provides several built-in methods to create arrays quickly:



**Zeros** : Creates an array filled with zeros of the specified size

*syntax* : np.zeros(shape)


```python
np.zeros(3)
```

**Ones** : Creates an array filled with ones of the specified size

*syntax* : np.ones(shape)


```python
np.ones(1)
```

Random:
Creates an array of random floats between 0 and 1

*syntax* : np.random.random(shape)

```python
np.random.random(3)
```

arange : Creates an array with a fixed step size betwen values

*syntax* : np.arange(start, stop, step)

start is inclusive

stop is exclusive

default value of step is 1


```python
np.arange(1,10)
```

```python
np.arange(1,10,2)
```

**linspace** : Creates an array of evenly spaced values between two numbers

*syntax* : np.linspace(start, stop, num)

start is inclusive

stop is inclusive

```python
np.linspace(1, 10, 5)
```

```python
np.linspace(1,10,10)
```
